```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('tested.csv')


df['Sex'] = df['Sex'].map({'male': 'Male', 'female': 'Female'})

survival_by_gender = df.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(6,4))
survival_by_gender.plot(kind='bar', color=['lightblue', 'pink'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.show()
```

```python
def display_survival_output(y_true, y_pred):
    for idx, (actual, predicted) in enumerate(zip(y_true, y_pred)):
        result = "SURVIVED" if predicted == 1 else "NOT SURVIVED"
        print(f"Passenger {idx+1}: Predicted = {result} | Actual = {'SURVIVED' if actual == 1 else 'NOT SURVIVED'} | UniqueCode: TITA-{idx+1:04d}-{result[:3]}")

display_survival_output(y_test.values, y_pred)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('tested.csv')

plt.figure(figsize=(6,4))
df['Survived'].value_counts().plot(kind='bar', color=['tomato', 'skyblue'])
plt.title('Titanic Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1], ['Not Survived', 'Survived'], rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('tested.csv')

survival_counts = df['Survived'].value_counts()
labels = ['Not Survived', 'Survived']
colors = ['orange', 'skyblue']


plt.figure(figsize=(6,6))
plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.05, 0.05))
plt.title('Titanic Survival Proportion')
plt.show()
```
